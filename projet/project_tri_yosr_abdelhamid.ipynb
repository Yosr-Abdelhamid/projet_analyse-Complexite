## <font color='blue' > Tri par Sélection </font> <br>  

```python
import random
def tableau_aléatoire(n):
    tab=[]
    for i in range(n):
        tab.append(random.randint(0, n))
    return tab 
```

```python
def tri_selection(tab):
    n = len(tab)
    for i in range (n):
      # Trouver le min
        min = i
        for j in range(i+1, len(tab)):
            if tab[min] > tab[j]:
                min = j
                
        tmp = tab[i]
        tab[i] = tab[min]
        tab[min] = tmp
    return tab
```

```python
import time
start_time = time.time()

n=20000
tab=tableau_aléatoire(n)
tri_selection(tab)
print("Le temps d'execution est : --- %s seconds ---" % (time.time() - start_time))

print (tab)
```

# <font color='blue' > Tri par Insertion</font> <br>   


```python
import random
def table_aléatoire(n):
    T=[]
    for i in range(n):
        T.append(random.randint(0, n))
    return T 
```

```python
def tri_insertion(T): 
    for i in range(1, len(T)): 
        k = T[i] 
        j = i-1
        while j >= 0 and k < T[j] : 
                T[j + 1] = T[j] 
                j -= 1
        T[j + 1] = k
    return(T)  
```

```python
import time
start_time = time.time()

n=20000
T=table_aléatoire(n)
tri_insertion(T)
print("Le temps d'execution est : --- %s seconds ---" % (time.time() - start_time))

print (T)
```

# <font color='blue' > Tri à bulle </font> <br>    

```python
import random
def tab_aléatoire(n):
    t=[]
    for i in range(n):
        t.append(random.randint(0, n))
    return t 
```

```python
def tri_bulle(t):
    n = len(t)
    for i in range(n):
        for j in range(0, n-i-1):
        # échanger si l'élément trouvé est plus grand que le suivant
            if t[j] > t[j+1] :
                t[j], t[j+1] = t[j+1], t[j]
    return(t)
```

```python
import time
start_time = time.time()

n=100000
t=tab_aléatoire(n)
tri_bulle(t)
print("Le temps d'execution est : --- %s seconds ---" % (time.time() - start_time))

print (t)
```

# <font color='blue' > Tri Rapide </font> <br>

```python
import random
def tabb_aléatoire(n):
    table=[]
    for i in range(n):
        table.append(random.randint(0, n))
    return table 
```

```python
def tri_rapide(table):
    
    def trirap(table, g, d):
        pivot = table[(g+d)//2]
        i = g
        j = d
        while True:
            while table[i]<pivot:
                i+=1
            while table[j]>pivot:
                j-=1
            if i>j:
                break
            if i<j:
                table[i], table[j] = table[j], table[i]
            i+=1
            j-=1
        if g<j:
            trirap(table,g,j)
        if i<d:
            trirap(table,i,d)
 
    g=0
    d=len(table)-1
    trirap(table,g,d)
```

```python
import time
start_time = time.time()

n=100000
table=tabb_aléatoire(n)
tri_rapide(table)
print("Le temps d'execution est : --- %s seconds ---" % (time.time() - start_time))

print (table)
```
